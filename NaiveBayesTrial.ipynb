```python
import pandas as pd
import numpy as np

import sklearn

from sklearn.naive_bayes import BernoulliNB
from sklearn.naive_bayes import GaussianNB
from sklearn.naive_bayes import MultinomialNB

from sklearn.model_selection import train_test_split

from sklearn import metrics
from sklearn.metrics import accuracy_score

from sklearn.linear_model import LinearRegression
```

```python
data = pd.read_csv("/Users/bishmdl76/Downloads/adult.csv")
```

```python
occupation_missing=[]
for i in range(len(data.occupation)):
    if data.occupation[i] == '?':
        occupation_missing.append(i)
```

```python
data.drop(occupation_missing,axis=0,inplace=True)
```

```python
data.reset_index(drop=True, inplace=True)
data
```

```python
native_country=[]
count1=-1
for i in data.native_country:
    count1=count1+1
    if i=="?":
        native_country.append(count1)
```

```python
data.drop(native_country,axis=0,inplace=True)
```

```python
race_dummies = pd.get_dummies(data.race)
data = pd.concat([data,race_dummies],axis='columns')
data = data.drop(['race','White'],axis='columns')
```

```python
def sex_c(x):
    if x=="Male":
        return 0
    else:
        return 1
```

```python
data.sex=data.sex.apply(sex_c)
```

```python
native_country_dummies = pd.get_dummies(data.native_country)
data = pd.concat([data,native_country_dummies],axis='columns')
data = data.drop(['native_country','United-States'],axis='columns')
```

```python
education_dummies = pd.get_dummies(data.education)
data = pd.concat([data,education_dummies],axis='columns')
data = data.drop(['10th','education'],axis='columns')
```

```python
marital_status_dummies = pd.get_dummies(data.marital_status)
data = pd.concat([data,marital_status_dummies],axis='columns')
data = data.drop(['Divorced','marital_status'],axis='columns')
```

```python
occupation_dummies = pd.get_dummies(data.occupation)
data = pd.concat([data,occupation_dummies],axis='columns')
data = data.drop(['Tech-support','occupation'],axis='columns')
```

```python
relationship_dummies = pd.get_dummies(data.relationship)
data = pd.concat([data,relationship_dummies],axis='columns')
data = data.drop(['Unmarried','relationship'],axis='columns')
```

```python
def numeric_income(x):
    if x=="<=50K":
        return 0
    else:
        return 1
```

```python
workclass_dummies = pd.get_dummies(data.workclass)
data = pd.concat([data,workclass_dummies],axis='columns')
data = data.drop(['Private','workclass'],axis='columns')
```

```python
data.income=data.income.apply(numeric_income)
```

```python
data
```

```python
X=data.drop(['income'],axis='columns')
```

```python
X
```

```python
Y=data.income
```

```python
Y
```

```python
X_train,X_test,Y_train,Y_test = train_test_split(X,Y,test_size=.25)
```

```python
Bernoulli=BernoulliNB(binarize=True)
Bernoulli.fit(X_train,Y_train)
print(Bernoulli)

Y_output=Bernoulli.predict(X_test)
print(accuracy_score(Y_test,Y_output))
```

```python
Gaussian=GaussianNB()
Gaussian.fit(X_train,Y_train)
print(Gaussian)

Y_output=Gaussian.predict(X_test)
print(accuracy_score(Y_test,Y_output))
```

```python
Multinomial=MultinomialNB()
Multinomial.fit(X_train,Y_train)
print(Multinomial)

Y_output=Multinomial.predict(X_test)
print(accuracy_score(Y_test,Y_output))
```
